```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd

plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams.update({'font.size':14})
plt.rcParams["figure.figsize"] = (10,6)

df = pd.read_csv('resultados_temporal.csv',encoding='utf-8',sep=',',index_col=[0])

parameters = ['Coliformes Termotolerantes', 'Oxigênio Dissolvido', 'Turbidez', 'Fósforo Total', 'Sólido Total', 'DBO (5, 20)', 'Temperatura da Água', 'pH']

for parameter in parameters:

    dfp = df.loc[df['parametro'] == parameter]

    g = sns.FacetGrid(dfp,col="parametro",height=7,aspect=0.5)
    g.map(sns.barplot, "tecnica","mape")
```
