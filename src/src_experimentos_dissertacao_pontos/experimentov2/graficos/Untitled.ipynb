```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams.update({'font.size':14})
plt.rcParams["figure.figsize"] = (10,6)

parameters = ['Coliformes Termotolerantes', 'Oxignio Dissolvido', 'Turbidez', 'Fsforo Total', 'Slido Total', 'DBO (5, 20)', 'Temperatura da gua', 'pH']

tecnicas = ['Regresso Linear', 'Random Forest', 'MLP', 'LSTM']

#df_resultado = pd.DataFrame(columns=['tecnica','parametro','amostras','mape','lag'])

df = pd.read_csv('../../experimentov2/espaco_temporal_pad_lag.csv',encoding='utf-8',sep=',',index_col=[0])

for parameter in parameters:
    
    for tecnica in tecnicas:
        
        if (tecnica == 'Regresso Linear'):
        
            df_RL = df[(df['parametro'] == parameter) & (df['tecnica'] == tecnica)]
           
            df_RL = df_RL.drop(['tecnica', 'parametro'], axis=1)
    
        elif (tecnica == 'Random Forest'):
        
            df_RF = df[(df['parametro'] == parameter) & (df['tecnica'] == tecnica)]
           
            df_RF = df_RF.drop(['tecnica', 'parametro'], axis=1)
            
        elif (tecnica == 'MLP'):
        
            df_MLP = df[(df['parametro'] == parameter) & (df['tecnica'] == tecnica)]
           
            df_MLP = df_MLP.drop(['tecnica', 'parametro'], axis=1)
        
        elif (tecnica == 'LSTM'):
        
            df_LSTM = df[(df['parametro'] == parameter) & (df['tecnica'] == tecnica)]
           
            df_LSTM = df_LSTM.drop(['tecnica', 'parametro'], axis=1)
        
    fig, ax = plt.subplots()
    ax.plot(df_RL.lag.values,df_RL.mape.values, label="Regresso Linear")
    ax.plot(df_RF.lag.values,df_RF.mape.values, label="Random Forest")
    ax.plot(df_MLP.lag.values,df_MLP.mape.values, label="MLP")
    ax.plot(df_LSTM.lag.values,df_LSTM.mape.values, label="LSTM")
    plt.title(parameter) #adicionando o ttulo
    plt.xlabel('Lags')
    plt.ylabel('MAPE(%)')
    ax.legend()
    plt.savefig('predicao_espaco_temporal_' + parameter + '.png')
    plt.show()
```
