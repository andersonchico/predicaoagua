```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 24
plt.rcParams.update({'font.size':20})
plt.rcParams["figure.figsize"] = (8,6)
```

```python

x = np.arange(3)
money = [2.14, 2.09, 2.28] # MAPE

width = 0.45  # the width of the bars
fig, ax = plt.subplots()
ax.set_ylabel('% Erro (MAPE)')
ax.set_xlabel('Modelos')
plt.bar(x, money,width)
plt.xticks(x, ('RL', 'MLP', 'LSTM'))
plt.show()
```

```python
x = np.arange(3)
money = [0.24, 0.24, 0.25] # RMSE

width = 0.45  # the width of the bars
fig, ax = plt.subplots()
ax.set_ylabel('Erro (RMSE)')
ax.set_xlabel('Modelos')
plt.bar(x, money,width)
plt.xticks(x, ('RL', 'MLP', 'LSTM'))
plt.show()
```

```python
#Comparao entre os modelos de regresso linear com 10%, 25% e 50% dos dados de treinamento como janela de visualizao.
#Metrica MAPE
labels = ['RL', 'MLP', 'LSTM'] #Modelos RL - Regresso Linear
men_means = [2.45, 2.18, 2.20] # Treinamento
women_means = [2.03, 1.93, 1.87] # teste

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Treinamento')
rects2 = ax.bar(x + width/2, women_means, width, label='Teste')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylim(0.6, 2.8)
ax.set_ylabel('% Erro (MAPE)')
ax.set_xlabel('Modelos')
#ax.set_title('RMSE Mdio por modelo e parmetro')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


#autolabel(rects1)
#autolabel(rects2)

fig.tight_layout()
#ax.legend(plt.legend(loc='upper left', bbox_to_anchor=(0.1, -0.4), shadow=True, ncol=8))
plt.show()
```
