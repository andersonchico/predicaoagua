```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['xtick.labelsize'] = 26
plt.rcParams['ytick.labelsize'] = 26
plt.rcParams.update({'font.size':26})
plt.rcParams["figure.figsize"] = (18,12)


labels = ['pH', 'OD', 'DBO', 'Fsforo', 'Turbidez']
men_means = [0.14, 0.17, 0.26, 0.01, 0.08]
women_means = [0.12, 0.11, 0.15, 0.00, 0.04]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='LSTM')
rects2 = ax.bar(x + width/2, women_means, width, label='MLP')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE Mdio')
ax.set_xlabel('Parmetros')
ax.set_title('RMSE Mdio por modelo e parmetro')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams.update({'font.size':20})
plt.rcParams["figure.figsize"] = (10,6)

data = {'pH': 0.02, 'OD': 0.03, 'DBO': 0.09, 'Fsforo': 0.00, 'Turbidez': 0.01}
names = list(data.keys())
values = list(data.values())

plt.plot(names, values,color='red',marker='o')
plt.ylabel('MSE mdio')
plt.xlabel('Parmetros')
plt.title('Modelo LSTM')
plt.show()

data2 = {'10': 0.036, '25': 0.034, '50': 0.035}
names2 = list(data2.keys())
values2 = list(data2.values())

plt.plot(names2, values2,marker='o')
plt.ylabel('MSE mdio')
plt.xlabel('Neurnios')
plt.title('Modelo LSTM')
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(10, 3), sharey=True)
axs[0].plot(names, values,color='red',marker='o')
axs[1].plot(names2, values2,marker='o')
fig.suptitle('Categorical Plotting')
```
