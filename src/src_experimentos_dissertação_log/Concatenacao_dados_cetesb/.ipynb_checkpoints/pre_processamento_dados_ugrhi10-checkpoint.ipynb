```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams.update({'font.size':16})
plt.rcParams["figure.figsize"] = (8,5)
```

```python
for parameter in ('ph','coliformes', 'dbo','fosforo','od','solido','temperatura','turbidez'):
        
    #print(parameter)
    # Carregando a base de dados:
    
    df = pd.read_csv(parameter + '.csv',encoding='utf-8',sep=';')[['data_coleta', 'valor','UGRHI']]
    
    df = df.loc[df['UGRHI'] == '10 - SOROCABA/MEDIO TIETE']

    df['data_coleta'] = pd.to_datetime(df['data_coleta'])
    df.index = df['data_coleta']
    df = df.groupby([pd.Grouper(freq='1M')]).mean().reset_index()
    df.columns = ['data_coleta',parameter]
    
    if (parameter == 'ph'):
        DFmerge = df
    else:
        DFmerge = pd.merge(DFmerge, df,how="outer", on="data_coleta") 
DFmerge
```

```python
df10 = DFmerge
df10=df10.set_index('data_coleta')
#Mostra os dados dos parÃ¢metros graficamente.
for c in df10.columns:
    plt.plot(df10[c])
    plt.title(c)
    plt.xlabel('Tempo')
    plt.ylabel('Valor ' + c)
    plt.show()
```

```python
#FunÃ§Ãµes que calculam os valores superior e inferior do boxplot. TambÃ©m calcula o valor mÃ©dio dos parÃ¢metros.
def outlier_superior(vparametro,df):
    #Ordena o dataframe
    trainingds = df.sort_values(vparametro,axis=0)

    tamanho = len(trainingds)
   
    #Calcula o 1Q, ou seja, os primeiros 25% do diagrama de caixa.
    primeiro_quartil = trainingds.quantile(q=0.25, axis=0, numeric_only=True, interpolation='linear')
    #Calcula o 3Q, ou seja, 75% do diagrama de caixa.
    terceiro_quartil = trainingds.quantile(q=0.75, axis=0, numeric_only=True, interpolation='linear')

    #Calcula-se o interquartil(FIQ).
    fig = terceiro_quartil - primeiro_quartil
    
    #Calcula o limite superior.
    superior = terceiro_quartil + (1.5*fig)
    
    return superior

def outlier_inferior(vparametro,df):
    #Ordena o dataframe
    trainingds = df.sort_values(vparametro,axis=0)

    tamanho = len(trainingds)
    
    #Calcula o 1Q, ou seja, os primeiros 25% do diagrama de caixa.
    primeiro_quartil = trainingds.quantile(q=0.25, axis=0, numeric_only=True, interpolation='linear')
    #Calcula o 3Q, ou seja, 75% do diagrama de caixa.
    terceiro_quartil = trainingds.quantile(q=0.75, axis=0, numeric_only=True, interpolation='linear')
      
    #Calcula-se o interquartil(FIQ).
    fig = terceiro_quartil - primeiro_quartil
    
    #Calcula o limite inferior.
    inferior = primeiro_quartil - (1.5*fig)
           
    return inferior

def media_coluna(vparametro,df):
    tamanho = len(df)
    
    soma = 0
    
    for i in range(tamanho):
        soma = soma + df[vparametro][i]
    
    #calcula o Valor medio da coluna
    media = soma / tamanho
        
    return media

def mediana_coluna(vparametro,df):
    
    mediana = df.median().vparametro
            
    return mediana
```

```python
#O parÃ¢metro OD nÃ£o necessita de tratamento de outlier por nÃ£o apresentar valores fora do padrÃ£o no boxplot da UGRHI 06
for i in ('ph','coliformes', 'dbo','fosforo','od','solido','temperatura','turbidez'):
    
    if (i == 'coliformes'):
        outs = outlier_superior(i,df10)[1]
        regs = len(df10[(df10.coliformes > outs)])
                
        outi = outlier_inferior(i,df10)[1]
        regi = len(df10[(df10.coliformes < outi)])
    
           
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[1],i] = df10.mean().coliformes
            outs = outlier_superior(i,df10)[1]
            regs = len(df10[(df10.coliformes > outs)]) 
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[1],i] = df10.mean().coliformes
            outi = outlier_inferior(i,df10)[1]
            regi = len(df10[(df10.coliformes < outi)])
            
        media = df10.mean().coliformes
                        
    elif(i == 'dbo'):
        outs = outlier_superior(i,df10)[2]
        regs = len(df10[(df10.dbo > outs)])
        
        outi = outlier_inferior(i,df10)[2]
        regi = len(df10[(df10.dbo < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[2],i] = df10.mean().dbo
            outs = outlier_superior(i,df10)[2]
            regs = len(df10[(df10.dbo > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[2],i] = df10.mean().dbo
            outi = outlier_inferior(i,df10)[2]
            regi = len(df10[(df10.dbo < outi)])
            
        media = df10.mean().dbo
                
    elif(i == 'fosforo'):
        
        outs = outlier_superior(i,df10)[3]
        regs = len(df10[(df10.fosforo > outs)])
        
        outi = outlier_inferior(i,df10)[3]
        regi = len(df10[(df10.fosforo < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[3],i] = df10.mean().fosforo
            outs = outlier_superior(i,df10)[3]
            regs = len(df10[(df10.fosforo > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[3],i] = df10.mean().fosforo
            outi = outlier_inferior(i,df10)[3]
            regi = len(df10[(df10.fosforo < outi)])
            
        media = df10.mean().fosforo
                
    elif(i == 'ph'):
        
        outs = outlier_superior(i,df10)[0]
        regs = len(df10[(df10.ph > outs)])
        
        outi = outlier_inferior(i,df10)[0]
        regi = len(df10[(df10.ph < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[0],i] = df10.mean().ph
            outs = outlier_superior(i,df10)[0]
            regs = len(df10[(df10.ph > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[0],i] = df10.mean().ph
            outi = outlier_inferior(i,df10)[0]
            regi = len(df10[(df10.ph < outi)])
            
        media = df10.mean().ph
                
    elif(i == 'solido'):
        
        outs = outlier_superior(i,df10)[5]
        regs = len(df10[(df10.solido > outs)])
        
        outi = outlier_inferior(i,df10)[5]
        regi = len(df10[(df10.solido < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[5],i] = df10.mean().solido
            outs = outlier_superior(i,df10)[5]
            regs = len(df10[(df10.solido > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[5],i] = df10.mean().solido
            outi = outlier_inferior(i,df10)[5]
            regi = len(df10[(df10.solido < outi)])
            
        media = df10.mean().solido
                
    elif(i == 'temperatura'):
        
        outs = outlier_superior(i,df10)[6]
        regs = len(df10[(df10.temperatura > outs)])
        
        outi = outlier_inferior(i,df10)[6]
        regi = len(df10[(df10.temperatura < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[6],i] = df10.mean().temperatura
            outs = outlier_superior(i,df10)[6]
            regs = len(df10[(df10.temperatura > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[6],i] = df10.mean().temperatura
            outi = outlier_inferior(i,df10)[6]
            regi = len(df10[(df10.temperatura < outi)])
            
        media = df10.mean().temperatura
        
    elif(i == 'od'):
        
        outs = outlier_superior(i,df10)[4]
        regs = len(df10[(df10.od > outs)])
        
        outi = outlier_inferior(i,df10)[4]
        regi = len(df10[(df10.od < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[4],i] = df10.mean().od
            outs = outlier_superior(i,df10)[4]
            regs = len(df10[(df10.od > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[4],i] = df10.mean().od
            outi = outlier_inferior(i,df10)[4]
            regi = len(df10[(df10.od < outi)])
            
        media = df10.mean().od
                        
    else:
        
        
        outs = outlier_superior(i,df10)[7]
        regs = len(df10[(df10.turbidez > outs)])
        
        outi = outlier_inferior(i,df10)[7]
        regi = len(df10[(df10.turbidez < outi)])
    
        while (regs > 0):
            df10.loc[df10[i] > outlier_superior(i,df10)[7],i] = df10.mean().turbidez
            outs = outlier_superior(i,df10)[7]
            regs = len(df10[(df10.turbidez > outs)])
        
        while (regi > 0):
            df10.loc[df10[i] < outlier_inferior(i,df10)[7],i] = df10.mean().turbidez
            outi = outlier_inferior(i,df10)[7]
            regi = len(df10[(df10.turbidez < outi)])
            
        media = df10.mean().turbidez
             

#Mostra os outlier dos valores de cada parÃ¢metro da UGRHI06 apÃ³s redimensionamento semanal
print('Dados da UGRHI 06')
#Mostra os dados por parÃ¢metro
for c in df10.columns:
    fig4,ax4 = plt.subplots()
    ax4.set_title(c)
    ax4.boxplot(df10[c])
```
