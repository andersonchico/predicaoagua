```python
import plotnine as p9
import pandas as pd
import matplotlib.pyplot as plt 
from plotnine import ggplot, aes, geom_line,labs,theme,geom_bar,geom_point,geom_col,theme_bw
```

```python
DFmerge = pd.DataFrame()
dfbaseline = pd.read_csv('metricas_temporal_baseline.csv',encoding='utf-8',sep=',')
DFmerge = pd.concat([DFmerge, dfbaseline])
dftecnicas = pd.read_csv('metricas_temporal_tecnicas.csv',encoding='utf-8',sep=',')
DFmerge = pd.concat([DFmerge, dftecnicas])
```

```python
##Verifica os parâmetros que possuem os melhores resultados.
DF = DFmerge

myplot = (p9.ggplot(DF, p9.aes('lag', 'mape', fill='tecnica', color='factor(tecnica)')) +
  p9.geom_line() +
  p9.facet_wrap('~parametro', scales='free_x') + labs(x="Lags",y="MAPE") + theme_bw())

#myplot.save(filename = 'fig_dbo2.png', height=5, width=3.2, units = 'in')

myplot
#DF
```

```python
## Verifica os parâmetros que possuem MAPE <= 15%
DF = DFmerge

#DF = DF[DF['parametro'] != 'Coliformes Termotolerantes']
##Seleciona os resultados com MAPE <= 15.
DF = DF[DF['mape'] <= 15] 
#Seleciona a média dos parâmetros por técnicas.
#DF = DF.groupby(['parametro','tecnica']).mean()

myplot = (p9.ggplot(DF, p9.aes('lag', 'mape', fill='tecnica', color='factor(tecnica)')) +
  p9.geom_line() +
  p9.facet_wrap('~parametro', scales='free_x') + labs(x="Lags",y="MAPE") + theme_bw() + theme(text=p9.element_text(size=16)))

myplot.save(filename = 'fig_4parametros.png', height=5, width=8, units = 'in')

myplot
#DF
```

```python
##Plotagem do gráfico da série temporal dos dados de teste: compara os valores observados com os preditos de cada técnica.
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams.update({'font.size':16})
plt.rcParams["figure.figsize"] = (6,4)

DFmerge = pd.DataFrame()
parameters = ['Sólido Total', 'DBO (5, 20)', 'Temperatura da Água', 'pH']

for parameter in parameters:
    
    DFmerge = pd.DataFrame()
    dfbaseline = pd.read_csv('serie_temporal_baseline_'+ parameter +'.csv',encoding='utf-8',sep=',')
    dfbaseline = dfbaseline[dfbaseline['lags'] == 2]
    dftecnicas = pd.read_csv('predicao_serie_temporal_'+ parameter +'.csv',encoding='utf-8',sep=',')
    dftecnicas = dftecnicas[dftecnicas['lags'] == 2]
    
    dfrl   = dftecnicas[dftecnicas['tecnica'] == 'Regressão Linear']
    dfrf   = dftecnicas[dftecnicas['tecnica'] == 'Random Forest']
    dfmlp  = dftecnicas[dftecnicas['tecnica'] == 'MLP']
    dflstm = dftecnicas[dftecnicas['tecnica'] == 'LSTM']
    
    plt.plot(dfbaseline.observado.values,label='Observado Baseline', c = 'DarkGoldenrod',linestyle = '--')
    plt.plot(dfbaseline.previsao.values,label='Previsto Baseline', c = 'black')
    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title(parameter)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
    
    plt.plot(dfrl.observado.values,label='Observado', c = 'DarkGoldenrod',linestyle = '--')
    plt.plot(dfrl.previsao.values,label='Previsto RL', c = 'black')
    plt.plot(dfrf.previsao.values,label='Previsto RF', c = 'red')
    plt.plot(dfmlp.previsao.values,label='Previsto MLP', c = 'blue')
    plt.plot(dflstm.previsao.values,label='Previsto LSTM', c = 'LimeGreen')
    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title(parameter)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()


    
```

```python
##Agrupa por parâmetro e técnica, e calcula a média dentre os resultados com MAPE <= 15%
DF = DFmerge

##Seleciona os resultados com MAPE <= 15.
DF = DF[DF['mape'] <= 15] 
#Seleciona a média dos parâmetros por técnicas.
DF = DF.groupby(['parametro','tecnica']).mean()
DF.round(2)
```

```python
##Agrupa por técnica, e calcula a média dentre os resultados com MAPE <= 15%
DF = DFmerge

##Seleciona os resultados com MAPE <= 15.
DF = DF[DF['mape'] <= 15] 
#Seleciona a média geral das técnicas.
DF = DF.groupby(['tecnica']).mean()
DF.round(2)
```

```python
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams.update({'font.size':16})
plt.rcParams["figure.figsize"] = (6,4)
##Agrupa por técnica, e calcula a média dentre os resultados com MAPE <= 15%
DF = DFmerge

##Seleciona os resultados com MAPE <= 15.
DF = DF[DF['mape'] <= 15] 
DF = DF[DF['tecnica'] != 'Baseline'] 
#Seleciona a média geral das técnicas.
DF = DF.groupby(['lag']).mean()
DF = DF.reset_index()
DF = DF.round(2)

plt.plot(DF.lag.values,DF.mape.values,linestyle = '--',marker = 'o')
plt.xlabel('Lags')
plt.ylabel('MAPE')
#plt.title('')
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('lag6_desempenho.png')
plt.show()

DF
```
